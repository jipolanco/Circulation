```julia
# To run with multiple threads, see https://github.com/JuliaLang/IJulia.jl#installing-additional-julia-kernels
using Base.Threads
println("Running on $(nthreads()) threads")

using Circulation
using GPFields

using LinearAlgebra

@time import PyPlot
const plt = PyPlot;
```

```julia
const DATA_DIR_BASE = let dir = expanduser("~/Dropbox/circulation/data")
    @show readdir(dir)
    joinpath(dir, "4vortices")
end

const DATA_IDX = 1
const RESOLUTION = 256

const GP_PARAMS = ParamsGP(
    (RESOLUTION, RESOLUTION),
    L = (2pi, 2pi),
    c = 1.0,
    nxi = 1.5,
)

let dir = joinpath(DATA_DIR_BASE, string(RESOLUTION))
    println("Directories in $DATA_DIR_BASE: ", readdir(DATA_DIR_BASE))
    println("Files in $dir: ", readdir(dir))
end
```

```julia
# Load data
psi = let datadir = joinpath(DATA_DIR_BASE, string(GP_PARAMS.dims[1]))
    @show datadir
    @time GPFields.load_psi(GP_PARAMS, datadir, DATA_IDX)
end
summary(psi)
```

```julia
rho = GPFields.density(psi)
summary(rho)
```

```julia
p = GPFields.momentum(psi, GP_PARAMS);
v = map(pj -> pj ./ rho, p);
vreg = map(pj -> pj ./ sqrt.(rho), p);
@show extrema(p[1]);
@show extrema(v[1]);
@show extrema(vreg[1]);
```

```julia
Ls = GP_PARAMS.L
Ip = IntegralField2D(p[1], L=Ls);
@time prepare!(Ip, v);
```

```julia
using BenchmarkTools

Ns = GP_PARAMS.dims
κ = GP_PARAMS.κ

# Make loop around (Nx/4, Ny/4)
I0 = Ns .>> 2  # N / 2^2 = N / 4
dI = Ns .>> 6  # N / 2^6 = N / 64
loop = Circulation.Rectangle(I0 .- dI, 2 .* dI)

loop = Circulation.Rectangle(I0 .- (1, 1), 2 .* (1, 1))

# In physical coordinates
loop_phys = (Ls ./ Ns) * loop

println(loop)
println(loop_phys)

circulation(loop, Ip) / κ
```

```julia
fig, ax = plt.subplots()
# mappable = ax.imshow(log10.(abs.(v[2])))
mappable = ax.imshow(v[2])
fig.colorbar(mappable, ax=ax)
kwargs = (color=:grey, lw=1.0, )
xc = Circulation.get_centre(loop)
rect = plt.matplotlib.patches.Rectangle(loop.x, loop.r...; fill=false, kwargs...)
ax.add_patch(rect)
ax.plot(xc...; marker=:x, markersize=5, kwargs...)  # rectangle centre
ax.set_xlim(xc[1] - loop.r[1], xc[1] + loop.r[1])
ax.set_ylim(xc[2] - loop.r[2], xc[2] + loop.r[2])
```

```julia
# Compute full circulation from regularised velocity
loop_length = 12
circ = similar(v[1]) :: Matrix
loop_dims = (loop_length, loop_length)

@time prepare!(Ip, vreg)  # compute integral field w(x, y)

circulation!(circ, Ip, loop_dims)   # precompilation (not needed, but for accurate timings...)
@time circulation!(circ, Ip, loop_dims);  # compute circulation

@show norm(circ)

# @btime circulation!($circ, $Ip, $loop_dims);
```

# Resampling

```julia
using FFTW

psi_f = fft(psi)
psi_fine = similar(psi_f, 2 .* size(psi_f));
GPFields.resample_field_fourier!(psi_fine, psi_f, GP_PARAMS)
ifft!(psi_fine)
```

```julia
rho_fine = GPFields.density(psi_fine)

v_fine = let rho = rho_fine
    p = GPFields.momentum(psi_fine, gp_params_fine);
    v = map(pj -> pj ./ rho, p);
    # vreg = map(pj -> pj ./ sqrt.(rho), p);
    @show extrema(p[1]);
    @show extrema(v[1]);
    @show extrema(vreg[1]);
    v
end;
```

```julia
gp_params_fine = ParamsGP(GP_PARAMS, dims=size(psi_fine))
xy = get_coordinates(GP_PARAMS)
xy_fine = get_coordinates(gp_params_fine)

fig, ax = plt.subplots(dpi=120)
ax.set_aspect(:equal)
cf = ax.contourf(xy..., rho')
ax.contour(xy_fine..., rho_fine', colors="black", linewidths=0.5)
fig.colorbar(cf, ax=ax)

# ax.quiver(xy..., v[1]', v[2]', scale=12)
ax.quiver(xy_fine..., v_fine[1]', v_fine[2]', scale=12)

figlims = (0.45pi, 0.55pi)
ax.set_xlim(figlims...)
ax.set_ylim(figlims...)

# plt.imshow(abs2.(psi_fine))
# plt.imshow(abs2.(psi))
# plt.colorbar()
```
